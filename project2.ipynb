```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df['Date'] = pd.to_datetime(df['Date'])


df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

df['Total Sales'] = pd.to_numeric(df['Total Sales'], errors='coerce')

df_pivot = df.pivot_table(values='Total Sales', index='Year', columns='Month', aggfunc='sum')


plt.figure(figsize=(12, 6))
sns.heatmap(df_pivot, cmap='Reds', annot=True, fmt=".0f", linewidths=0.5)

# ÿ™ÿ≠ÿ≥ŸäŸÜ ÿßŸÑÿπÿ±ÿ∂
plt.title("Seasonal Sales Heatmap: Identifying High Sales Periods")
plt.xlabel("Month")
plt.ylabel("Year")
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
# ÿßŸÑÿ£ŸÑŸàÿßŸÜ ÿßŸÑÿØÿßŸÉŸÜÿ© ÿ™ÿπŸÜŸä ÿßÿ±ÿ™ŸÅÿßÿπ ÿßŸÑŸÖÿ®Ÿäÿπÿßÿ™ÿå ÿ®ŸäŸÜŸÖÿß ÿßŸÑÿ£ŸÑŸàÿßŸÜ ÿßŸÑŸÅÿßÿ™ÿ≠ÿ© ÿ™ÿπŸÜŸä ÿßŸÜÿÆŸÅÿßÿ∂Ÿáÿß
```
