في القطاع الصحي


```python
import matplotlib.pyplot as plt

df['Date'] = pd.to_datetime(df['Date'])


plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Health Transactions'], marker='o', linestyle='-', label="Health Transactions", color='blue')
plt.plot(df['Date'], df['Health Sales'], marker='s', linestyle='--', label="Health Sales", color='red')

plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Trend of Health Transactions and Sales Over Time")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
# تحليل نمو قطاع الصحة عبر الزمن
# الفكرة: معرفة ما إذا كان هناك ارتفاع أو انخفاض في عدد المعاملات والمبيعات في قطاع الصحة بمرور الوقت.
```

```python

df['Health Transactions'] = pd.to_numeric(df['Health Transactions'], errors='coerce')


plt.figure(figsize=(10, 5))
df.boxplot(column='Health Transactions', by='Month', grid=False)
plt.xlabel("Month")
plt.ylabel("Number of Transactions")
plt.title("Monthly Distribution of Health Transactions")
plt.suptitle("")
plt.show()
# تحليل تأثير الموسمية على الطلب الصحي
# الفكرة: هل يزداد الطلب على الخدمات الصحية في أشهر معينة (مثل الشتاء بسبب الأمراض الموسمية؟
```

```python
Q1 = df['Health Transactions'].quantile(0.25)
Q3 = df['Health Transactions'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Health Transactions'] < lower_bound) | (df['Health Transactions'] > upper_bound)]
print(outliers)
#outliers
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["Health Transactions"], df["Health Sales"], alpha=0.5, color='purple')
plt.xlabel("Health Transactions")
plt.ylabel("Health Sales")
plt.title("Relationship Between Health Transactions and Sales")
plt.grid(True)
plt.show()
# العلاقة بين عدد المعاملات والمبيعات في قطاع الصحة
# الفكرة: هل زيادة عدد العمليات تعني زيادة المبيعات، أم أن هناك عوامل أخرى تؤثر على المبيعات؟
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df['Date'] = pd.to_datetime(df['Date'])


df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

df['Total Sales'] = pd.to_numeric(df['Total Sales'], errors='coerce')

df_pivot = df.pivot_table(values='Total Sales', index='Year', columns='Month', aggfunc='sum')


plt.figure(figsize=(12, 6))
sns.heatmap(df_pivot, cmap='Reds', annot=True, fmt=".0f", linewidths=0.5)

# تحسين العرض
plt.title("Seasonal Sales Heatmap: Identifying High Sales Periods")
plt.xlabel("Month")
plt.ylabel("Year")
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
# الألوان الداكنة تعني ارتفاع المبيعات، بينما الألوان الفاتحة تعني انخفاضها
```
